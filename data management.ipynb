```python
from scipy.stats import ttest_ind

# Esempio di t-test per le differenze di genere nelle posizioni informatiche nel 2022
male_ict = df[(df['sex'] == 'M') & (df['TIME_PERIOD'] == 2022)]['ict_people']
female_ict = df[(df['sex'] == 'F') & (df['TIME_PERIOD'] == 2022)]['ict_people']

t_stat, p_value = ttest_ind(male_ict, female_ict)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_value}")
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Esempio di regressione lineare per prevedere il numero di persone in posizioni informatiche
X = df[['TIME_PERIOD']]
y = df['ict_people']

# Creazione del modello
model = LinearRegression()
model.fit(X, y)

future_year = 2023
predicted_ict_people = model.predict(np.array([[future_year]]))
print(f"Previsto numero di persone in posizioni informatiche nel {future_year}: {predicted_ict_people[0]}")
```

```python
anni_futuri = np.arange(2023, 2027)
previsioni_ict = model.predict(anni_futuri.reshape(-1, 1))
previsioni_df = pd.DataFrame({'Anno': anni_futuri, 'Previsto_ict_people': previsioni_ict})
print(previsioni_df)
```

#### Nei dataset precedenti l'anno più recente è il 2022, mentre in questo è il 2021, tramite il metodo della regressione lineare vado a creare gli ultimi dati che mi servono per riportare il tutto su dashboard, per questo metodo utlizzo i commenti nel codice affinchè possa essere seguito semplicemente

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

# Ottenego la lista delle nazioni univoche nel tuo dataset
nazioni_univoche = internet['geo'].unique()

# Inizializzo un DataFrame vuoto per le previsioni finali
previsioni_finali = pd.DataFrame()

# Itero attraverso le diverse nazioni
for nazione in nazioni_univoche:
    # Filtro i dati solo per la nazione corrente
    dati_nazione = internet[internet['geo'] == nazione]

    # Seleziono solo i dati fino al 2021 per questa nazione
    dati_addestramento = dati_nazione[dati_nazione['TIME_PERIOD'] <= 2021]

    # Creo un modello di regressione lineare 
    model = LinearRegression()
    model.fit(dati_addestramento[['TIME_PERIOD']], dati_addestramento['OBS_VALUE'])

    # Crea un DataFrame con i dati per il 2022 
    dati_previsione_2022 = pd.DataFrame({'geo': [nazione], 'TIME_PERIOD': [2022]})

    # Effettuo la previsione per il 2022 
    previsione_2022 = model.predict(dati_previsione_2022[['TIME_PERIOD']])

    # Aggiungo le previsioni per questa nazione al DataFrame finale
    dati_previsione_2022['OBS_VALUE'] = previsione_2022
    previsioni_finali = pd.concat([previsioni_finali, dati_previsione_2022], ignore_index=True)
```

```python
# Unisco il DataFrame delle previsioni con il DataFrame originale
df_con_anno_2022 = pd.concat([internet, previsioni_finali], ignore_index=False)
```
